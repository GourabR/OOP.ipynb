```python
from bs4 import BeautifulSoup
import requests
import re
import pandas as pd
```

```python
url = 'https://www.primevideo.com/storefront/ref=atv_hm_pri_c_9zZ8D2_hm_mv?contentType=movie&contentId=home'
user_agent = 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/94.0.4606.81 Safari/537.36'
headers = ({'user_agent':user_agent,'Accept-Language':'en-US, en;q=0.5'})
webpage = requests.get(url, headers = headers)
webpage
```

```python
soup = BeautifulSoup(webpage.content, 'html.parser')
print(soup.prettify)
```

```python
links = soup.find_all("div", attrs = {'class':'g-box eros-box movie-g-box'})
```

```python
print(links)
```

```python
movies = []

for link in links:
  genre_link = link.find("a", class_="g-box-content-wrap").get("href")
  new_url = "https://www.erosnow.com"+genre_link
  user_agent = 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/94.0.4606.81 Safari/537.36'
  headers = ({'user_agent':user_agent,'Accept-Language':'en-US, en;q=0.5'})
  new_webpage = requests.get(new_url, headers = headers)
  new_soup = BeautifulSoup(new_webpage.content, 'html.parser')
  new_links = new_soup.find_all("a", attrs = {'class':'m-box-content'})
  for link in new_links:
    movie_url = link.get("href")
    movie_url = "https://www.erosnow.com"+movie_url
    another_webpage = requests.get(movie_url, headers = headers)
    another_soup = BeautifulSoup(another_webpage.content, 'html.parser')
    if another_soup.find("a", attrs = {'class':'entity-sub-heading-link router-link-exact-active router-link-active'}) is not None:
      movie_name = another_soup.find("a", attrs = {'class':'entity-sub-heading-link router-link-exact-active router-link-active'}).text
    else:
      movie_name = None
    if another_soup.find("div", attrs = {'class':'extra-content-item actors-wrap'}) is not None:
      actors = another_soup.find("div", attrs = {'class':'extra-content-item actors-wrap'}).text
    else:
      actors = None
    if another_soup.find("div", attrs = {'class':'extra-content-item directors-wrap'}) is not None:
      directors = another_soup.find("div", attrs = {'class':'extra-content-item directors-wrap'}).text
    else:
      directors = None
    if another_soup.find_all("div", attrs = {'class':'extra-content-item lang-wrap'})[0] is not None:
      language = another_soup.find_all("div", attrs = {'class':'extra-content-item lang-wrap'})[0].text
    else:
      language = None
    if another_soup.find_all("div", attrs = {'class':'extra-content-item lang-wrap'})[1] is not None:
      genre = another_soup.find_all("div", attrs = {'class':'extra-content-item lang-wrap'})[1].text
    else:
      genre = None
    if another_soup.find("div", attrs = {'class':'extra-info-content-rating'}) is not None:
      rating = another_soup.find("div", attrs = {'class':'extra-info-content-rating'}).text
    else:
      rating = None
    if another_soup.find("a", attrs = {'class':'entity-release-link'}) is not None:
      released_year = another_soup.find("a", attrs = {'class':'entity-release-link'}).text
    else:
      released_year = None
    if another_soup.find("div", attrs = {'class':'overview-content'}) is not None:
      Overview = another_soup.find("div", attrs = {'class':'overview-content'}).text
    else:
      Overview = None

    movies.append({
        'movie_name':movie_name,
        'Actors':actors,
        'Directors':directors,
        'Language':language,
        'Genre':genre,
        'Rating':rating,
        'Released_year':released_year,
        'Movie_overview':Overview
    })


df = pd.DataFrame(movies)
```
