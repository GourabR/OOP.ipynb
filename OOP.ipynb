```python
import sqlite3 as sql

con = sql.connect('sigma.db')
cur = con.cursor()

cur.execute('''CREATE TABLE IF NOT EXISTS my_database
            (id INTEGER PRIMARY KEY AUTOINCREMENT,
            name VARCHAR(255),
            email VARCHAR(255),
            password VARCHAR(255))''')
```

```python
# Creating a class
class BikeRental:

# A constructor in which some variables are initiated.
  def __init__(self):
    self.name = ''
    self.email = ''
    self.password = ''
    self.pulsar = 20
    self.apache = 15
    self.R15 = 13
    self.amount1 = 0
    self.amount2 = 0
    self.amount3 = 0
    self.pulsar_quantity = 0
    self.apache_quantity = 0
    self.R15_quantity = 0
    self.pulsar_hours = 0
    self.apache_hours = 0
    self.R15_hours = 0

    user_name = input("Please enter your name :")
    print(f"Welcome {user_name} to www.bike.rental.com")
    self.menu()

# MenuOptions: Providing menu options for the user to choose from.
  def menu(self):
    user_input = input('''How would you like to proceed?
    press 1. To check bikes availabe
    press 2. To check price
    press 3. To book bikes
    Press 4. For any query
    Press 5. To Exit
    ''')
    if user_input == '5':
      print("Thank you for visiting our website!")

    if user_input == '1':
      self.bikes_availabe()

    if user_input == '2':
      self.check_price()
    
    if user_input == '3':
      self.book_bikes()

    if user_input == '4':
      self.query()

# BikesAvailable: Displaying the available bikes to the user.
  def bikes_availabe(self):
    print('''We have 3 types of bikes availabe.
    1.Pulsar
    2.Apache
    3.R15''')
    self.menu()

# CheckPrice: Showing the price of each bike per hour.
  def check_price(self):
    print('''
    Pulsar Rs.70/hour
    Apache Rs.60/hour
    R15 Rs. 90/hour''')
    self.menu()

# BookBikes: Handling the booking process of bikes based on user input.
  def book_bikes(self):
    try:
        self.registration()
        user_input = input('''Which bike would you like to book
        Pulsar
        Apache
        R15'''
        )
        if user_input == 'Pulsar' or user_input == 'pulsar':
          print("We have {} quantity.".format(self.pulsar))
          self.pulsar_quantity = int(input("How many Pulsar would you like to book?"))
          while self.pulsar_quantity > self.pulsar:
            print(f"We don't have more than {self.pulsar} Pulsar at this moment. However we are trying to increase the number of vehicals.")
            self.pulsar_quantity = int(input("Please enter how many Pulsar would you like to book?"))
          else:
            self.pulsar_hours = int(input("Please enter for how many hours would you like to ride pulsar"))
            self.amount1 = self.pulsar_quantity*self.pulsar_hours*70
            self.another_bike()

        if user_input == 'Apache' or user_input == 'apache':
          print("We have {} quantity.".format(self.apache))
          self.apache_quantity = int(input("Please enter how many Apache would you like to book?"))
          while self.apache_quantity > self.apache:
            print(f"We don't have more than {self.apache} Apache at this moment. However we are trying to increase the number of vehicals.")
            self.apache_quantity = int(input("Please enter how many Apache would you like to book?"))
          else:
            self.apache_hours = int(input("Please enter for how many hours would you like to ride Apache"))
            self.amount2 = self.apache_quantity*self.apache_hours*60
            self.another_bike()

        if user_input == 'R15' or user_input == 'r15':
          print("We have {} quantity.".format(self.R15))
          self.R15_quantity = int(input("Please enter how many bikes would you like to book?"))
          while self.R15_quantity > self.R15:
            print(f"We don't have more than {self.R15} quantities of R15 at this moment. However we are trying to increase the number of vehicals.")
            self.R15_quantity = int(input("Please enter how many R15 would you like to book?"))
          else:
            self.R15_hours = int(input("Please enter for how many hours would you like to ride R15"))
            self.amount3 = self.R15_quantity*self.R15_hours*90
            self.another_bike()
    except Exception as e:
      print("An error occurred during bike booking",str(e))
      self.menu()
        

# AnotherBike: Handling the booking of additional bikes if requested by the user.
  def another_bike(self):
    any_bike = input("Any other bike would you like to book? (yes/no)")
    if any_bike.lower() == 'yes':
      again = input("Which bike would you like to book?")
      if again == 'Apache':
        print(f"We have {self.apache} quantities")
        self.apache_quantity = int(input("How many Apache would you like to book?"))
        while self.apache_quantity > self.apache:
          print(f"We don't have more than {self.apache} of Apache at this moment. However we are trying to increase the number of vehicals.")
          self.apache_quantity = int(input("Please enter how many Apache would you like to book?"))
        else:
          self.apache_hours = int(input("Please enter for how many hours would you like to ride apache"))
          self.amount2 = self.apache_quantity*self.apache_hours*60
          self.another_bike()

      if again == 'R15':
        print(f"We have {self.R15} quantities")
        self.R15_quantity = int(input("How many R15 would you like to book?"))
        while self.R15_quantity > self.R15:
          print(f"We don't have more than {self.R15} of R15 at this moment. However we are trying to increase the number of vehicals.")
          self.R15_quantity = int(input("Please enter how many R15 would you like to book?"))
        else:
          self.R15_hours = int(input("Please enter for how many hours would you like to ride R15"))
          self.amount3 = self.R15_quantity*self.R15_hours*90
          self.another_bike()

      if again == 'Pulsar':
        print(f"We have {self.pulsar} quantities")
        self.pulsar_quantity = int(input("How many Pulsar would you like to book?"))
        while self.pulsar_quantity > self.pulsar:
          print(f"We don't have more than {self.pulsar} of pulsar at this moment. However we are trying to increase the number of vehicals.")
          self.pulsar_quantity = int(input("Please enter how many pulsar would you like to book?"))
        else:
          self.pulsar_hours = int(input("Please enter for how many hours would you like to ride pulsar"))
          self.amount1 = self.pulsar_quantity*self.pulsar_hours*90
          self.another_bike()
    else:
      self.payment()
      print("Enjoy your ride!")

# Payment: Managing the payment process for the booked bikes.
  def payment(self):
    try:
        self.amount = self.amount1 + self.amount2 + self.amount3
        self.total_bikes = self.apache_quantity + self.pulsar_quantity + self.R15_quantity
        self.total_hours = self.apache_hours + self.pulsar_hours + self.R15_hours
        print(f"The payable amount for total {self.total_bikes} bikes for {self.total_hours} hours is Rs.{self.amount}")
        payment_input = input("Please select a payment mode :")
        while payment_input != 'Gpay' and payment_input != 'Debit card':
          print("We only accept Gpay and Debit card as payemnt mode.")
          payment_input = input("Please select a payment mode :")
        else:
          if payment_input == 'Gpay':
            user_input = input("Please type your UPI ID here :")
            print("Your payment is successful")
          else:
            user_input = input("Please type your card number here :")
            print("Your payment is successful!")
    except Exception as e:
      print('''It seems that there was an issue with the payment.
              If your money has been deducted from your bank account, rest assured that it will be credited back to your account within 2 business days.''', str(e))

# Query: Handling user queries and providing assistance.
  def query(self):
    try:
        self.registration()
        user_input = input("Please enter your query here :")
        query_id = self.generate_query_id()
        query_details = {
                'query_id': query_id,
                'user_name': self.name,
                'email': self.email,
                'query': user_input,
                'status': 'Submitted'
            }
        print(f"Your query has been submitted. Your query ID is {query_id}.")
        print("We understand your concern, our customer representative will get back to you shortly!")
        print("Thank you for visiting our website!")
        self.save_query(query_details)
    except Exception as e:
      print("An error occurred. Please try again", str(e))
      self.menu()

# Generating some random no for query_id.
  def generate_query_id(self):
    import random
    random_no = random.randint(1000,10000)

    return random_no

# Saving the query for future reference.
  def save_query(self,query_details):
    self.query_details = query_details


# Registration: Managing the user registration process.
  def registration(self):
      user_name = input("Please enter your full name for registration: ")
      self.name = user_name
      user_email = input("Please enter your email: ")

      while '@gmail.com' not in user_email:
          print("This email is not valid")
          user_email = input("Please enter a valid email: ")
      else:
          self.email = user_email

      cur.execute('''SELECT email FROM my_database''')
      email_list = cur.fetchall()
      email_list = [row[0] for row in email_list]

      if self.email in email_list:
          print("A profile with this email id already exists")
          user_input = input("Do you wish to continue with the same email id? (Yes/No): ")
          if user_input.lower() == 'yes':
              self.login()
          else:
              self.registration()
      else:
          self.password = input("Please enter a password: ")
          while len(self.password) < 8 or not any(char.islower() for char in self.password) or not any(
                  char.isupper() for char in self.password) or not any(
              char.isdigit() for char in self.password) or not any(
              char in '!@#$%^&*()_-+={}[]:;"\'|<>,.?/~`' for char in self.password):
              print(
                  "Your password should contain at least 8 characters (1 lowercase alphabet, 1 uppercase alphabet, 1 numeric value, and 1 special character)")
              self.password = input("Please enter a password: ")

          cur.execute('''INSERT INTO my_database (name, email, password) VALUES (?, ?, ?)''', (self.name, self.email, self.password))
          print("Registration successful!")
          con.commit()

# Login: Creating a login method
  def login(self):
    try:
        user_email = input("Please enter your email for login:")
        while '@gmail.com' not in user_email:
            print("This email is not valid")
            user_email = input("Please enter a valid email: ")
        else:
            self.email = user_email
            cur.execute(f'''SELECT password FROM my_database WHERE email = ?''', (self.email))
            result = cur.fetchone()
            if result:
                db_password = result[0]
                user_password = input("please enter your password for login:")
                while user_password != db_password:
                    print("Incorrect password!")
                    user_password = input("please enter your password for login:")
                else:
                    self.password = user_password
                    print("Login successful!")
            else:
                print("Email not found in the database. Please register.")
    except Exception as e:
      print("Something went worng. Please try again", str(e))

# Creating an object of "BikeRental" class.
bj = BikeRental()
```
